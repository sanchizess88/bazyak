```python
import math

# Данные
n = 30          # количество бросков
p = 0.75        # вероятность попадания
q = 1 - p       # вероятность промаха

# Функция для вероятности ровно k попаданий
def P(k):
    return math.comb(n, k) * (p**k) * (q**(n-k))

# а) Ровно 25 попаданий
p_a = P(25)
print(f"а) Ровно 25: {p_a:.6f}")

# б) не менее 25 попаданий (т.е. 25,26,27,28,29,30)
p_b = 0
for k in range(25, 31):
    p_b += P(k)
print(f"б) Не менее 25: {p_b:.6f}")

# в) менее 25 попаданий (т.е. 0..24)
p_v = 0
for k in range(0, 25):
    p_v += P(k)
print(f"в) Менее 25: {p_v:.6f}")

# г) более 25 попаданий (т.е. 26..30)
p_g = 0
for k in range(26, 31):
    p_g += P(k)
print(f"г) Более 25: {p_g:.6f}")

# д) не более 25 попаданий (т.е. 0..25)
p_d = 0
for k in range(0, 26):
    p_d += P(k)
print(f"д) Не более 25: {p_d:.6f}")

# е) ровно 30 попаданий
p_e = P(30)
print(f"е) Ровно 30: {p_e:.6f}")
```
