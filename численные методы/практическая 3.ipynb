## Задание 1
Методом итераций вычислите один корень заданного уравнения с точностью 10-3
с помощью «ручной» расчетной таблицы и калькулятора.

```python
import math


x = 1.1
eps = 0.001  # точность
tau = 0.3  # параметр шага для обеспечения сходимости (метод релаксации)

print(f"{'n':<4} | {'x_n':<10} | {'|x_new - x|':<12}")
print("-" * 35)

n = 0
while True:
    # f(x) - левая часть уравнения, приведенного к виду f(x) = 0
    f_x = x * math.sin(x) - 1

    # Итерационная формула: x_{n+1} = x_n - tau * f(x_n)
    x_new = x - tau * f_x

    diff = abs(x_new - x)
    print(f"{n:<4} | {x:<10.5f} | {diff:<12.5f}")

    if diff < eps:
        x = x_new
        break

    x = x_new
    n += 1

print("-" * 35)
print(f'Корень: {x:.4f}')

# Проверка: подставляем корень в исходное уравнение
left_side = x * math.sin(x)
right_side = 1
print(f'Проверка: {left_side:.4f} = {right_side:.4f}')
```
