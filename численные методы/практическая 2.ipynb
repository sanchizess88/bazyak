## Задание 1
Отделите корни заданного уравнения, пользуясь графическим методом.

```python
import math
import numpy as np
import matplotlib.pyplot as plt


def f(x):
    """Функция f(x) = x * sin(x) - 1"""
    return x * np.sin(x) - 1

# Для наглядного отделения корней представим уравнение как x * sin(x) = 1
x_vals = np.linspace(0, 10, 1000)  # Рассмотрим интервал от 0 до 10
y1 = x_vals * np.sin(x_vals)
y2 = np.ones_like(x_vals)  # y = 1 (правая часть уравнения)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y1, label='y = x·sin(x)', color='blue', linewidth=2)
plt.plot(x_vals, y2, label='y = 1', color='red', linestyle='--', linewidth=2)
plt.grid(True)
plt.legend()
plt.title('Отделение корней уравнения x·sin(x) = 1 (Вариант 7)')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2, 10)
plt.axhline(y=0, color='black', linewidth=0.5)
plt.axvline(x=0, color='black', linewidth=0.5)
plt.show()

# Поиск интервалов, где функция f(x) меняет знак
print("\nПоиск интервалов с корнями:")
for i in range(len(x_vals)-1):
    if f(x_vals[i]) * f(x_vals[i+1]) < 0:
        print(f"Корень находится на отрезке [{x_vals[i]:.2f}; {x_vals[i+1]:.2f}]")
```


## Задание 2
По методу половинного деления вычислите один корень заданного уравнения с
точностью 10-3 с помощью «ручной» расчетной таблицы и калькулятора.

```python
import math
import numpy as np
import matplotlib.pyplot as plt


def f_scalar(x):
    """Функция для скалярных вычислений (без numpy)"""
    return x * math.sin(x) - 1

# Начальное приближение выбираем из графика (первый положительный корень ~1.1)
x_curr = 1.1
eps = 0.001  # Точность
tau = 0.3    # Параметр релаксации (подбирается для обеспечения сходимости)

print(f"{'n':<4} | {'x_n':<12} | {'f(x_n)':<12} | {'|x_next - x_n|':<14}")
print("-" * 55)

n = 1
while True:
    f_val = f_scalar(x_curr)
    x_next = x_curr - tau * f_val
    diff = abs(x_next - x_curr)

    print(f"{n:<4d} | {x_curr:<12.6f} | {f_val:<12.6f} | {diff:<14.6f}")

    if diff < eps:
        break

    x_curr = x_next
    n += 1

print("-" * 55)
print(f"Найденный корень: x = {x_next:.6f}")
print(f"Проверка: f({x_next:.6f}) = {x_next * math.sin(x_next) - 1:.6f}")
print(f"Аналитическая проверка: sqrt(4x+7) = 3*cos(x) → {math.sqrt(4*x_next + 7):.6f} = {3*math.cos(x_next):.6f}")
```
