### Задание E

```python
import random  # Нужен для случайной фразы

class Parrot:
    def __init__(self, phrase):
        self.phrases = [phrase]      # Создаём список фраз и добавляем первую фразу
    
    def say(self, repeat=1):
        random_phrase = random.choice(self.phrases)
        print((random_phrase + " ") * repeat)       # Выбираем случайную фразу из списка и печатаем кол-во раз
    
    def learn(self, new_phrase):
        self.phrases.append(new_phrase)             # Добавляем новую фразу в список


# Пример
p = Parrot("Гав!")
p.say()
p.learn("Мяу!") 
p.say() 
p.say(3) 
```

### Задание А

```python
class LampRow:
    def __init__(self):
        self.__state = "00000000"       # Лампочки изначально выключены
    
    @property
    def state(self):
        return self.__state
    
    @state.setter
    def state(self, new_state):
        if len(new_state) == 8:     # Проверка, что строка состоит из 8 символов
            self.__state = new_state
        else:
            print("Ошибка! Длина строки должна быть 8. Устанавливаю все нули.")
            self.__state = "00000000"
    
    def show(self):
        # Показываем лампочки: 0 = выкл (-), 1 = вкл (*)
        result = ""
        for lamp in self.__state:
            if lamp == "0":
                result += "-"
            else:
                result += "*"
        print(result)

# Пример
lamps = LampRow()
lamps.show()                    # --------
lamps.state = "10101010"
print(lamps.state)              # 10101010
lamps.show()                    # *-*-*-*-
```

```python
class LampRow:
    def __init__(self, count=8):
        self.__count = count    # Запоминает количество лампочек
        self.__state = "0" * count  # Запоминает количество лампочек
    
    @property
    def state(self):
        return self.__state
    
    @state.setter
    def state(self, new_state):
        if len(new_state) == self.__count:  # Проверка длины строки
            # Проверка, что все символы - 0 или 1
            correct = True
            for c in new_state:
                if c != "0" and c != "1":
                    correct = False
                    break
            
            if correct:
                self.__state = new_state
            else:
                print("Ошибка! Нужно использовать только 0 и 1")
                self.__state = "0" * self.__count
        else:
            print(f"Ошибка! Нужно ровно {self.__count} символов")
            self.__state = "0" * self.__count
    
    def show(self):
        # Показываем лампочки: 0 -> '-', 1 -> '*'
        result = ""
        for lamp in self.__state:
            if lamp == "0":
                result += "-"
            else:
                result += "*"
        print(result)

# Пример
lamps = LampRow(6)        # создаём ряд из 6 лампочек
lamps.show()              # выведет: ------
lamps.state = "101010"
print(lamps.state)        # выведет: 101010
lamps.show()              # выведет: *-*-*-
lamps.state = "10101010"  # ошибка! (нужно 6 символов)
print(lamps.state)        # выведет: 000000
lamps.show()              # выведет: ------
```

### Задание С

```python
class LampRow:
    def __init__(self, count=8):
        self.__count = count
        self.__state = "0" * count
    
    @property
    def state(self):
        return self.__state
    
    @state.setter
    def state(self, new_state):
        # Проверка длины строки
        if len(new_state) != self.__count:
            print(f"Ошибка! Нужно ровно {self.__count} символов")
            self.__state = "0" * self.__count
            return
        
        for c in new_state:     # Проверка, что все символы - 0, 1 или 2
            if c not in "012":
                print("Ошибка! Можно использовать только 0, 1 или 2")
                self.__state = "0" * self.__count
                return
        
        # Если всё правильно
        self.__state = new_state
    
    def show(self):
        # Показываем лампочки:
        # 0 -> '-' (выключена)
        # 1 -> '*' (красная)
        # 2 -> 'o' (зелёная)
        result = ""
        for lamp in self.__state:
            if lamp == "0":
                result += "-"
            elif lamp == "1":
                result += "*"
            else:  # lamp == "2"
                result += "o"
        print(result)

# Пример
lamps = LampRow(6)
lamps.show()                    # выведет: ------
lamps.state = "102102"
print(lamps.state)              # выведет: 102102
lamps.show()                    # выведет: *-o*-o
lamps.state = "10201010"        # ошибка! (нужно 6 символов)
print(lamps.state)              # выведет: 000000
lamps.show()                    # выведет: ------
```

```python
class LampRow:
    def __init__(self, count=8):
        self.__count = count
        self.__state_int = 0    # Хранит состояние как целое число (0 - все выключены)
    
    @property
    def state(self):
        # Преобразуем целое число в строку при чтении И переводим в троичную систему и добавляем нули слева
        result = ""
        temp = self.__state_int
        for i in range(self.__count):
            # Получаем последнюю цифру в троичной системе
            digit = temp % 3    
            result = str(digit) + result
            temp //= 3
        
        return result
    
    @state.setter
    def state(self, new_state):
                # Проверка длины строки
        if len(new_state) != self.__count:
            print(f"Ошибка! Нужно ровно {self.__count} символов")
            self.__state_int = 0
            return
        
        for c in new_state:             # Проверка, что все символы - 0, 1 или 2
            if c not in "012":
                print("Ошибка! Можно использовать только 0, 1 или 2")
                self.__state_int = 0
                return
        
        # Преобразует строку в целое число (троичная система)
        value = 0
        for c in new_state:
            value = value * 3 + int(c)
        
        self.__state_int = value
    
    def show(self):
        # Получает строку состояния через свойство state
        state_str = self.state
        
        # Показывает лампочки
        for lamp in state_str:
            if lamp == "0":
                print("-", end="")
            elif lamp == "1":
                print("*", end="")
            else:  # lamp == "2"
                print("o", end="")
        print()

# Пример
lamps = LampRow(6)
lamps.show()                    # ------
lamps.state = "102102"
print(lamps.state)              # 102102
lamps.show()                    # *-o*-o
lamps.state = "10201010"        # ошибка! (нужно 6 символов)
print(lamps.state)              # 000000
lamps.show()                    # ------
```
